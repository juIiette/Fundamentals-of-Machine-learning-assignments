```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('KIEM.xlsx')
df1 = pd.read_excel('KIEM.xlsx')
```

```python
df = df.iloc[:, 6:31]
df1 = df1.iloc[:, 0:1]
df = pd.concat([df1, df], axis =1 )
```
